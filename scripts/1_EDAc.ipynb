```python
# ðŸ“¦ Import Packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
pd.set_option('display.float_format', lambda x: '%.2f' % x)

# ðŸ“¥ Load Dataset
df = pd.read_csv("/content/sample_data/kc_house_data.csv")


# ðŸ§ Initial Overview
print("ðŸ“Œ Dataset shape:", df.shape)
print("\nðŸ“Œ Data types:")
print(df.dtypes)
print("\nðŸ“Œ Missing values:")
print(df.isnull().sum())
print("\nðŸ“Œ First few rows:")
display(df.head())
print("\nðŸ“Œ Descriptive statistics:")
display(df.describe())

# ðŸ§¼ CLEANING
df = df.drop_duplicates()
df = df[(df['bedrooms'] > 0) & (df['bathrooms'] > 0)]

print("\nâœ… Cleaning Completed")
print("Remaining missing values:")
print(df.isnull().sum())
print("\n--- Cleaning Conclusion ---")
print("- Removed duplicates")
print("- Removed rows with 0 bedrooms or 0 bathrooms")
print("- No missing values found")

# ðŸ“Š UNIVARIATE ANALYSIS
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()

# ðŸ”„ BIVARIATE ANALYSIS


if 'id' in df.columns:
    df = df.drop(columns=['id'])
cols = df.columns.tolist()
if 'price' in cols:
    cols.insert(0, cols.pop(cols.index('price')))
    df = df[cols]

corr_matrix = df.corr(numeric_only=True)


plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, linewidths=0.5, cbar_kws={"shrink": .8})
plt.title("ðŸ“ˆ Pearson Correlation Matrix (ID Removed, Price First)", fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()



# ðŸ” MULTIVARIATE ANALYSIS
features = ['sqft_living', 'grade', 'bathrooms', 'sqft_above', 'view', 'floors']
for feat in features:
    plt.figure(figsize=(8,5))
    sns.scatterplot(x=df[feat], y=df['price'])
    plt.title(f"Price vs {feat}")
    plt.xlabel(feat)
    plt.ylabel("Price")
    plt.show()

# âš ï¸ OUTLIERS
plt.figure(figsize=(10,5))
sns.boxplot(x=df['price'])
plt.title("Boxplot - Price")
plt.show()

plt.figure(figsize=(10,5))
sns.boxplot(x=df['bedrooms'])
plt.title("Boxplot - Bedrooms")
plt.show()

# ðŸ—ºï¸ MAPPING
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="long", y="lat", hue="price", palette="viridis", alpha=0.6)
plt.title("House Locations (colored by price)")
plt.show()
```
